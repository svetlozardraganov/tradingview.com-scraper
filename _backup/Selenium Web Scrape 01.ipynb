```python
#SCRAPE FINANCIAL PARAMETERS FROM INDIVIDUAL PAGES

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import pandas as pd

chrome_options = Options()
chrome_options.add_argument("--window-size=1000,1080")
chrome_options.add_argument("--headless")

#find advertise window button and click it
#button = driver.find_element_by_xpath('//button[@class="Button__StyledButton-a1qza5-0 cUAUIG"]')
#button.click()

base_url = "https://www.macrotrends.net/stocks/charts/AAPL/apple/"

pages = ['revenue', 'gross-profit', 'operating-income', 'ebitda', 'net-income', 'eps-earnings-per-share-diluted', 'shares-outstanding',
         'total-assets', 'cash-on-hand', 'long-term-debt', 'total-liabilities', 'total-share-holder-equity',
         'profit-margins', 'gross-margin', 'operating-margin', 'ebitda-margin', 'pre-tax-profit-margin', 'net-profit-margin',
         'pe-ratio', 'price-sales', 'price-book', 'price-fcf', 'net-worth']

income_statement_urls = ['income_statement_urls', 'revenue', 'cost-goods-sold', 'gross-profit', 'research-development-expenses',
                        'selling-general-administrative-expenses', 'operating-expenses', 'operating-income',
                        'total-non-operating-income-expense', 'pre-tax-income', 'total-provision-income-taxes',
                        'income-after-taxes', 'income-from-continuous-operations', 'net-income', 
                        'ebitda', 'ebit', 'basic-shares-outstanding', 'shares-outstanding', 
                        'eps-basic-net-earnings-per-share', 'eps-earnings-per-share-diluted']

ballance_sheet_urls = ['ballance_sheet_urls', 'cash-on-hand', 'receivables-total', 'inventory', 'other-current-assets', 
                      'total-current-assets', 'net-property-plant-equipment', 'long-term-investments', 
                      'goodwill-intangible-assets-total', 'other-long-term-assets', 'total-long-term-assets', 
                      'total-assets', 'total-current-liabilities', 'long-term-debt', 'other-non-current-liabilities', 
                      'total-long-term-liabilities', 'total-liabilities', 'common-stock-net', 'retained-earnings-accumulated-deficit',
                      'comprehensive-income', 'total-share-holder-equity', 'total-liabilities-share-holders-equity']

cashflow_statement_urls = ['cashflow_statement_urls', 'net-income-loss', 'total-depreciation-amortization-cash-flow', 'other-non-cash-items',
                          'total-non-cash-items', 'change-in-accounts-receivable', 'change-in-inventories',
                          'change-in-accounts-payable', 'change-in-assets-liabilities', 'total-change-in-assets-liabilities', 
                          'cash-flow-from-operating-activities' , 'net-change-in-property-plant-equipment', 'net-change-in-intangible-assets',
                          'net-acquisitions-divestitures', 'net-change-in-short-term-investments', 'net-change-in-long-term-investments', 
                          'net-change-in-investments-total', 'investing-activities-other', 'cash-flow-from-investing-activities', 
                          'net-long-term-debt', 'net-current-debt', 'debt-issuance-retirement-net-total', 'net-common-equity-issued-repurchased', 
                          'net-total-equity-issued-repurchased', 'total-common-preferred-stock-dividends-paid', 'financial-activities-other', 
                          'cash-flow-from-financial-activities', 'net-cash-flow', 'stock-based-compensation', 'common-stock-dividends-paid']

common_ratios_urls = ['common_ratios_urls', 'current-ratio', 'long-term-debt-capital', 'debt-equity-ratio', 'gross-margin', 
                     'operating-margin', 'ebit-margin', 'ebitda-margin', 'pre-tax-profit-margin', 
                     'net-profit-margin', 'asset-turnover', 'inventory-turnover', 'receiveable-turnover', 
                     'days-sales-in-receivables', 'roe', 'return-on-tangible-equity', 'roa', 'roi', 'book-value-per-share', 
                     'operating-cash-flow-per-share', 'free-cash-flow-per-share',
                     'pe-ratio', 'price-sales', 'price-book', 'price-fcf', 'net-worth', 'quick-ratio']

other_statements_urls = ['other_statements_urls', 'dividend-yield-history', 'number-of-employees', 'stock-price-history', 'stock-splits',
                        'market-cap']

#eps-earnings-per-share-diluted, profit-margins, gross-margin, operating-margin, ebitda-margin, pre-tax-profit-margin, net-profit-margin
#pe-ratio, price-sales, price-book, price-fcf, net-worth
#//*[@id="style-1"]/div[2]/table/tbody/tr

#i = 4
#url = base_url + pages[i]
#print(pages[i])
#driver.get(url)
#
#table = driver.find_element_by_xpath('//*[@id="style-1"]/div[1]/table/tbody')
#table_name = driver.find_element_by_xpath('//*[@id="style-1"]/div[1]/table/thead')
#print(table_name.text)
#
#table_split = table.text.splitlines()
#table_years = []
#table_values = []
#for i in range(len(table_split)):
#    line_split = table_split[i].split()
#    table_years.append(line_split[0])
#    table_values.append(line_split[1])
#    
#
#df = pd.DataFrame([table_values], columns=table_years, index=[pages[i]])
#df = df.append(df)


def scrape_data(base_url, statement_urls):
    mult = 22
    print('<'*mult, "start-scraping", '>'*mult)
    
    driver = webdriver.Chrome("G:\My Drive\Investing\Programming\chromedriver.exe", options=chrome_options)
    #//div[@id="contenttablejqxgrid"]/div[1]/div[3] - 
    driver.implicitly_wait(5)
    company_data = pd.DataFrame()
    
        
    if statement_urls[0] == 'income_statement_urls' or statement_urls[0] == 'ballance_sheet_urls':
        annual_xpath = '//*[@id="style-1"]/div[1]/table/tbody' #annual report table
    elif statement_urls[0] == 'cashflow_statement_urls':
        annual_xpath = '//*[@id="style-1"]/div[2]/table/tbody' #annual report table
    elif statement_urls[0] == 'common_ratios_urls':
        annual_xpath = '//*[@id="style-1"]/table/tbody' #annual report table
        

    #iterate over company statements
    for i in range(1, len(statement_urls)):
        #vars to store values in / clear vars for every iteration
        data = None
        table_years = []
        table_values = []

        try:
            url = base_url + statement_urls[i]
            driver.get(url)
            table = driver.find_element_by_xpath(annual_xpath) 
                                               
            table_split = table.text.splitlines() #split table to lines

            #iterate over table-lines
            for j in range(len(table_split)): 
                line_split = table_split[j].split() #split line to year and value
                table_years.append(line_split[0]) #store years
                table_values.append(line_split[1]) #store values

            data = pd.DataFrame([table_values], columns=table_years, index=[statement_urls[i]]) #create temp dataframe with scraped data
            #company_data = company_data.append(data)
            company_data = pd.concat([company_data, data]) #append temp-dataframe row to global dataframe
            print(statement_urls[i], " > done")
        except:
            print(statement_urls[i], " > failed!")
            
    print('<'*mult, "end-scraping", '>'*mult)
    driver.close()
    return company_data

#income_statement_urls [ok] ballance_sheet_urls [ok] cashflow_statement_urls [ok] common_ratios_urls [ok] other_statements_urls
apple_statements = scrape_data(base_url, common_ratios_urls)

```

```python
#SCRAPE MULTIPLE FINANCIAL PARAMETERS FROM SINGLE PAGE / V2 / Faster scrolling using drag&drop apprach
#https://stackoverflow.com/questions/62119348/how-to-scroll-horizontally-using-selenium-chromedriver-in-python

from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.action_chains import ActionChains

import pandas as pd
import time

import matplotlib.pyplot as plt

start = time.time()

chrome_options = Options()
#chrome_options.add_argument("--window-size=1000,1080")
#chrome_options.add_argument("--headless")

driver = webdriver.Chrome("G:\My Drive\Investing\Programming\chromedriver.exe", options=chrome_options)
driver.implicitly_wait(5)
driver.maximize_window()
    
url = "https://www.macrotrends.net/stocks/charts/AAPL/apple/income-statement?freq=Q"
driver.get(url)

#find advertise window button and click it
ad_button = driver.find_element_by_xpath('//button[@class="Button__StyledButton-a1qza5-0 cUAUIG"]')
ad_button.click()

#scroll window down
# driver.execute_script("window.scrollTo(100,400);")

horizontal_bar_width = driver.find_element_by_id('jqxScrollOuterWraphorizontalScrollBarjqxgrid').rect['width']
slider = driver.find_element_by_id('jqxScrollThumbhorizontalScrollBarjqxgrid')

previous_last_element = -1
current_last_element = 0
previous_last_element1 = -1
current_last_element1 = 0

periods_table_data = []
main_table_data = []
    
company_data = pd.DataFrame()

#raise SystemExit("Stop right there!")

while previous_last_element != current_last_element:
    previous_last_element = current_last_element
    previous_last_element1 = current_last_element1
    
    # periods_table = driver.find_element_by_xpath('//*[@id="columntablejqxgrid"]') #this is the header row in the table, containing the dates.
    periods_table = driver.find_element_by_xpath('//*[@id="contentjqxgrid"]/div[1]')
    
    #parameters_table = driver.find_element_by_xpath('//*[@id="contentjqxgrid"]/div[2]')
    periods_table_split = periods_table.text.splitlines() #split the periods to array
    current_last_element = periods_table_split[-1] #get the last element from the table
    # print(periods_table.text)
    # print('last-element={}'.format(current_last_element))
    
    # main_table = driver.find_elements_by_xpath('//div[@role="row"]') #get the main table
    main_table = driver.find_elements_by_xpath('//*[@id="contentjqxgrid"]/div[2]/div/div') #get the main table
    current_last_element1 = main_table[0].text.splitlines()[-1]
    
    if previous_last_element == 0: #this is the innitial position of the table, no scrolling
        periods_table_data.extend(periods_table_split) #add all elements to the table
        
        for i in range (len(main_table)): #add all parameter to table
            main_table_split = main_table[i].text.splitlines()
            main_table_data.append(main_table_split)
            
    else:
        previous_last_element_index = periods_table_split.index(previous_last_element) #find the previous last element when the table is scrolled to the right
        periods_table_data.extend(periods_table_split[previous_last_element_index+1:]) #add the new elements after the table is scrolled to the right
        
        previous_last_element_index1 = main_table[0].text.splitlines().index(previous_last_element1)
        
        for row in range (len(main_table)):
            main_table_split = main_table[row].text.splitlines()
            main_table_data[row].extend(main_table_split[previous_last_element_index1+1:])
        
        
        # raise SystemExit("Stop right there!")
    
    #add data to pandas dataframe
    
    # Ensure the slider is in view
    slider.location_once_scrolled_into_view
    #scroll table to the right
    ActionChains(driver).click_and_hold(slider).move_by_offset(horizontal_bar_width/5, 0).release().perform()
    

for i in range (len(main_table_data)):
    data = pd.DataFrame([main_table_data[i][1:]], columns=periods_table_data[1:], index=[main_table_data[i][0]]) #create temp dataframe with scraped data
    company_data = pd.concat([company_data, data]) #append temp-dataframe row to global dataframe


driver.close()
# print(periods_table_data)
end = time.time()
print('Complete time: {}s'.format(end - start))
```

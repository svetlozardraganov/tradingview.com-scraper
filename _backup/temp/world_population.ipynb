```python
import matplotlib.pyplot as plt

# load population data
population = [2.53, 7.57, 33.02, 65.18, 150.69, 513.52, 1379.06, 3611.32]
years = [1800, 1850, 1900, 1950, 2000, 2050, 2100, 2150]

# create a line chart
plt.plot(years, population)

# add labels and title
plt.xlabel("Year")
plt.ylabel("Population (in billions)")
plt.title("World Population")

# show the chart
plt.show()
```
